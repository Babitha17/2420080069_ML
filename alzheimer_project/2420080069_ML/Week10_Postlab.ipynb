```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
import warnings
warnings.filterwarnings("ignore")

# Transaction dataset
transactions = [
    ['Bread','Milk','Eggs'],
    ['Bread','Butter','Jam'],
    ['Milk','Bread','Butter'],
    ['Bread','Milk','Butter','Eggs'],
    ['Milk','Eggs','Cereal'],
    ['Bread','Eggs','Milk'],
    ['Butter','Jam','Bread'],
    ['Bread','Milk','Butter'],
    ['Milk','Cereal','Eggs'],
    ['Bread','Butter','Jam','Milk']
]

# Convert to binary matrix
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

print("Binary Transaction Matrix")
print(df)

# ---------------- APRIORI ----------------
print("\nAPRIORI FREQUENT ITEMSETS")

apriori_sets = apriori(df, min_support=0.3, use_colnames=True)

print(apriori_sets)

apriori_rules = association_rules(apriori_sets, metric="confidence", min_threshold=0.5)

print("\nApriori Rules")
print(apriori_rules[['antecedents','consequents','support','confidence','lift']])

# ---------------- FP GROWTH ----------------
print("\nFP-GROWTH FREQUENT ITEMSETS")

fp_sets = fpgrowth(df, min_support=0.3, use_colnames=True)

print(fp_sets)

fp_rules = association_rules(fp_sets, metric="confidence", min_threshold=0.5)

print("\nFP-Growth Rules")
print(fp_rules[['antecedents','consequents','support','confidence','lift']])

# ---------------- COMPARISON ----------------
print("\nComparison")

print("Apriori Itemsets:", len(apriori_sets))
print("FP-Growth Itemsets:", len(fp_sets))

print("\nObservation:")
print("FP-Growth is faster because it avoids candidate generation used in Apriori.")
```
